# Split dataset into train and test (70:30) 

```python
X=df.drop(['Selector'],axis=1)
y=df['Selector']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

# Fit a base model and explain the reason of selecting that model. Please write your key observations  

```python
## Decision tree 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score
dt=DecisionTreeClassifier(random_state=42)
dt.fit(X_train,y_train)
y_pred=dt.predict(X_test)
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
##Logistic Regression
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=42)
logreg.fit(X_train,y_train)
print(classification_report(y_test, logreg.predict(X_test)))
```

# How do you improve the accuracy of the model? Write clearly the changes that you will make before re-fitting the model. Fit the final model. 

```python
from imblearn.over_sampling import SMOTE
X_res,y_res=SMOTE().fit_resample(X_train,y_train)
dt_sm=DecisionTreeClassifier(random_state=42)
dt_sm.fit(X_res,y_res)
pred=dt_sm.predict(X_test)
print(classification_report(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
st=RandomForestClassifier(random_state=42,criterion='entropy')
st.fit(X_res,y_res)
ypredict=st.predict(X_test)
print(classification_report(y_test,ypredict))
```

```python
##Hybrid sampling technique
from imblearn.combine import SMOTETomek
X_res, y_res=SMOTETomek().fit_resample(X_train,y_train)
dt_sm=DecisionTreeClassifier(random_state=42)
dt_sm.fit(X_res, y_res)
pred=dt_sm.predict(X_test)
print(classification_report(y_test, pred))
```
